```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams['text.usetex'] = True
```

```python
x = np.linspace(-1,1,1000)
K = np.array([1, 2, 3, 5, 10, 20])
for k in K:
    y = np.cosh(k * np.arccosh(x+0j))
    plt.figure(), plt.plot(x, y.real, color='black'), plt.ylabel('$\mathcal{T}(\lambda)$'), plt.xlabel('$\lambda$'), plt.title('Cheby-' + str(k))
```
